```python
#import libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Direct pi calculation: SINGLE RUN

```python
import random

n_trials = 4000
n_hits = 0
for iteration in range(n_trials):
    x,y = random.uniform(-1,1), random.uniform(-1,1)
    if x**2 + y**2 < 1:
        n_hits +=1
print(4 * n_hits/n_trials)
```

Direct pi calculation: MULTIPLE RUN

```python
import random

def direct_pi(N):
    n_hits = 0
    for iteration in range(N): 
        x,y = random.uniform(-1,1), random.uniform(-1,1)
        if x**2 + y**2 < 1:
            n_hits += 1
    return n_hits
n_runs = 5
n_trials = 10000
for run in range(n_runs):
    print(4 * direct_pi(n_trials)/float(n_trials))
```

Pi Calculation Class:

```python
import random

class PeeblePi(object):
    
    def direct_pi(self, N):
        
        """
        N  - number of times peebles thrown
        n_runs - number of experiments repeated
        
        """ 
        n_hits = 0
        for iteration in range(N): 
            x,y = random.uniform(-1,1), random.uniform(-1,1)
            if x**2 + y**2 < 1:
                n_hits += 1
        return n_hits
    
obj_pi = PeeblePi()
n_runs = 5
N_trials = 10000

for run in range(n_runs):
    print(4*obj_pi.direct_pi(N_trials)/N_trials)
```

MARKOV-CHAIN SAMPLING: ADULTS IN HELIPORT

```python
def markov_pi_multi(N, delta): 
    x, y = 1.0, 1.0
    n_hits = 0
    for i in range(N):
        del_x, del_y = random.uniform(-delta, delta), random.uniform(-delta, delta)
        if abs(x + del_x) < 1.0 and abs(y + del_y) < 1.0:
            x, y = x + del_x, y + del_y
        if x**2 + y**2 < 1.0: n_hits += 1
    return n_hits
n_runs = 50
n_trials = 1000
delta = 0.1
for n_trial in range(100,n_trials, 200):
    pi_list = []
    for run in range(n_runs):
        pi = 4 * markov_pi_multi(n_trial, delta)/n_trial
        pi_list.append(pi)
    plt.figure(figsize =(10,6))
    sns.lineplot(data = pi_list, label = 'Pi')
    
    plt.xlabel('runs')
    plt.ylabel('Pi_value')
    #plt.show()
```


# pebble_basic_inhomogeneous.py

```python
import random
import numpy as np

#initializing histogram
histo = [0, 0, 0, 0, 0, 0, 0, 0, 0]
# 3x3 peeble game set up
neighbor =  [[1, 3, 0, 0], [2, 4, 0, 1], [2, 5, 1, 2],
             [4, 6, 3, 0], [5, 7, 3, 1], [5, 8, 4, 2],
             [7, 6, 6, 3], [8, 7, 6, 4], [8, 8, 7, 5]]

#defining inhomogeneous weights
weight = [3.0, 0.5, 1.0, 0.5, 1.0, 0.5, 2.0, 0.5, 1.0] #inhomogeneous weight

pos = 8
n_iter = 10000000
for iter in range(n_iter):
    new_pos = neighbor[pos][random.randint(0, 3)]
    if random.random() < weight[new_pos] / weight[pos]:
        pos = new_pos
    histo[pos] += 1
norm = sum(weight)
print('comparison: weight,     histogram')
cal_weight = []
print(len(weight))
for k in range(len(weight)):
    cal_weight.append(norm * histo[k] / float(n_iter))
    print('site: ', k,' weight: ', weight[k], ' histo: ', norm * histo[k] / float(n_iter))
print(cal_weight)
```

```python
import pandas as pd
# making dataframe of calculated weight. given weight, and difference for plots
df = pd.DataFrame(list(zip(cal_weight, weight)), 
               columns =['cal_weight', 'weight']) 
df['difference'] = df['cal_weight'] - df['weight']
df
```

# Direct sampling of pi: root mean square deviation vs number of trials

```python
import random, math, pylab
def direct_pi(N): 
    n_hits = 0
    for i in range(N):
        x, y = random.uniform(-1.0, 1.0), random.uniform(-1.0, 1.0)
        if x ** 2 + y ** 2 < 1.0:
            n_hits += 1
    return n_hits
n_runs = 500
n_trials_list = []
sigmasqs = []
for poweroftwo in range(4, 13): 
    n_trials = 2 ** poweroftwo
    sigmasq = 0.0
    for run in range(n_runs):
        pi_est = 4 * direct_pi(n_trials) / float(n_trials)
        
        sigmasq += (pi_est - math.pi) **2
    sigmasqs.append(math.sqrt(sigmasq/(n_runs)))
    n_trials_list.append(n_trials)
pylab.plot(n_trials_list, sigmasqs, 'o')
pylab.xscale('log')
pylab.yscale('log')
pylab.xlabel('number of trials')
pylab.ylabel('root mean square deviation')
pylab.title('Direct sampling of pi: root mean square deviation vs n_trials')
pylab.savefig('direct_sampling_rms_deviation.png')
pylab.plot([10.0, 100.0], [1.642 / math.sqrt(10.0), 1.642 / math.sqrt(100.0)])
pylab.savefig('direct_sampling_rms_deviation_fit.png')
pylab.show()
```

# markov_sampling_rms_deviation

```python
import random, math, pylab

def markov_pi(N, delta):
    x, y = 1.0, 1.0
    n_hits = 0
    for i in range(N):
        del_x, del_y = random.uniform(-delta, delta), random.uniform(-delta, delta)
        if abs(x + del_x) < 1.0 and abs(y + del_y) < 1.0:
            x, y = x + del_x, y + del_y
        if x ** 2 + y ** 2 < 1.0: n_hits += 1
    return n_hits

n_runs = 500
delta_list = [0.062, 0.125, 0.25, 0.5, 1.0, 2.0, 4.0]
#delta_list = [0.062,3,1,2,0.5]
plt.figure(figsize =(12,6))
for delta in delta_list:
    n_trials_list = []
    sigmas = []
    for poweroftwo in range(4, 13):
        n_trials = 2 ** poweroftwo
        sigma = 0
        for run in range(n_runs):
            pi_est = 4.0 * markov_pi(n_trials, delta) / float(n_trials)
            sigma += (pi_est - math.pi) ** 2
            
        sigmas.append(math.sqrt(sigma/(n_runs)))
        n_trials_list.append(n_trials)
    
    pylab.plot(n_trials_list, sigmas, 'o', ms = 8, label = '$\delta = $' + str(delta))

pylab.xscale('log')
pylab.yscale('log')
pylab.xlabel('number of trials')
pylab.ylabel('root mean square deviation')
pylab.plot([10,10000],[1.642 / math.sqrt(10.0), 1.642 / math.sqrt(10000.0)], label = 'direct')
pylab.title('Markov-chain sampling of pi: root mean square deviation vs. n_trials')

pylab.legend(loc='upper right')
pylab.savefig('markov_sampling_rms_deviation.png')
pylab.show()
```

# bunching method

```python
import math, random
n_trials = 20000000
n_hits = 0
var = 0
for iter in range(n_trials):
    x, y = random.uniform(-1,1), random.uniform(-1,1)
    obs = 0
    if x**2 + y**2 < 1:
        n_hits += 1
        obs = 4
    var += (obs - math.pi)**2
print(4 * n_hits/n_trials, math.sqrt(var/ n_trials))
```
